```python
import pandas as pd

# Replace this with your file path
file_path = r"C:\Users\mouc074\OneDrive - PNNL\Documents\wind_energy_research\wind_data\HRRR_data\56774.csv"

# Load the CSV file
df = pd.read_csv(file_path)


df['valid_time'] = pd.to_datetime(df['valid_time'])
df = df.sort_values('valid_time').reset_index(drop=True)

full_time_index = pd.date_range(start=df['valid_time'].min(),
                                end=df['valid_time'].max(),
                                freq='H')
df = df.set_index('valid_time')
df = df.reindex(full_time_index)
df.index.name = 'valid_time'
station_name = df['station'].dropna().iloc[0] 
df_numeric = df.drop(columns='station')
df_numeric = df_numeric.interpolate(method='linear')
df_numeric['station'] = station_name
df = df_numeric.reset_index()

save_path = r"C:\Users\mouc074\OneDrive - PNNL\Documents\wind_energy_research\wind_data\HRRR_data\interpolated\56774.csv"
df.to_csv(save_path, index=False)
```
